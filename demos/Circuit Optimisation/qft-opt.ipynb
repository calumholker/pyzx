```python
import sys, os, time
sys.path.append('../..')
import pyzx as zx
import numpy as np
```

```python
c_i = zx.Circuit.load("../../circuits/benchmarking_circuits/Fast/before/QFT8")
zx.draw(c_i)
print(f'2Q-count = {c_i.twoqubitcount()}')
print(f'Non-Clifford gate count = {c_i.tcount()}')
```

```python
g = c_i.to_graph()
zx.to_graph_like(g)
zx.teleport_reduce(g)
zx.id_fuse_simp(g, quiet=True)
c_f = zx.extract_circuit(g)
c_f = zx.basic_optimization(c_f)
zx.draw(c_f)
print(f'2Q-count = {c_f.twoqubitcount()}')
print(f'Non-Clifford gate count = {c_f.tcount()}')
```

These two circuits are equal as evidenced by equal matrices. Note that the way the matrices are calculated introduces some rounding errors (e.g. see the 0,0 element below).

```python
c_i.to_matrix()[0,0] - c_f.to_matrix()[0,0]
```

```python
np.allclose(c_i.to_matrix(), c_f.to_matrix(), atol=1e-10)
```
